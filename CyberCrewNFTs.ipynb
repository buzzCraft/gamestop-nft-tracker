```python
from numpy import dot
#@title NTF last 100 trades
#@markdown Just click the arrow to start

#@markdown Timezone : GMT - 30 minute lag due to the API
dot_marker = True #@param {type:"boolean"}
CloneCard = True #@param {type:"boolean"}
CyberClone = True #@param {type:"boolean"}
CyberCycle = True #@param {type:"boolean"}
Cand = True #@param {type:"boolean"}
TenWorld = False #@param {type:"boolean"}
Cannon = False #@param {type:"boolean"}
LoadingLvl = False #@param {type:"boolean"}


# GNU GENERAL PUBLIC LICENSE

# Will show the 100 latest trades on different NFT's
# [CYBΞR BUZZΞ]




import requests
import pandas as pd
from datetime import datetime
import plotly.express as px

CARD = ["0x27665297fab3c72a472f81e6a734ffe81c8c1940a82164aca76476ca2b506724", "CLONE CARD"]
CLONE = ["0x2c4b4edd628bcffffbf4a9d434ce83e4737889b65eee922f00d3c2b2d82b3394", "CYBER CLONE"]
CYCLE = ["0x057047417d4aaf63a083ed0b379d8b8d44f7a9edf6252dced73be6147928eaaf", "CYBER CYCLE"]
CAND = ["0x230d40e35852948fe84d3a4077aefb3c1ae11297b94a55bafc9c8fc1793585ca", "CAN-D"]
TENWOR = ["0x20c7f321f7d800f38f3fb62fd89cbfc28072feea226c0bc9bde0efc2ce008f01", "10-Worlds"]
CANNON = ["0x09eb5d265456b098fa29ca27a63da34a3756d888838cbc8f17e4ccda256adcd3", "CHROME CANNON"]
LOADING = ["0x0d1a4f4d19f4aaaaf01cfc1eee2c24294653ab376fb0daf319ae6fdb2063c4a3", "LOADING LEVEL"]
items = []
# items = [CARD,CLONE,CYCLE,CAND]
if CloneCard:
  items.append(CARD)
if CyberClone:
  items.append(CLONE)
if CyberCycle:
  items.append(CYCLE)
if Cand:
  items.append(CAND)
if TenWorld:
  items.append(TENWOR)
if Cannon:
  items.append(CANNON)
if LoadingLvl:
  items.append(LOADING)
if len(items)==0:
  pass


def get_data(token_id, name, data_list = []):

  URL = "https://api.nft.gamestop.com/nft-svc-marketplace/history?nftData="
  response = requests.get(URL+token_id)
  # response.raise_for_status()
  # access JSOn content
  jsonResponse = response.json()
  for block in jsonResponse:
      for key, value in block.items():
          if key == "createdAt":
              ts = int(value)/1000.0 # Because its a Unix epoch with milliseconds
              # created = datetime.fromtimestamp(ts)
              created = datetime.utcfromtimestamp(ts)

          if key == "transaction":
              for k, v in value.items():
                  if k == "orderB":
                    qty = int(v.get("amountS"))
                    paid = int(v.get("amountB"))/1e18
                    paid = paid/qty
                      
                    data_list.append([created,paid,qty, name])
  return data_list

lst = []
for nft in items:
  lst = get_data(nft[0], nft[1])




df = pd.DataFrame(lst, columns =['Time GMT', 'Price ETH', 'Quanty', 'Name']) 
fig = px.line(df, x = "Time GMT", y = "Price ETH", color = 'Name', markers=dot_marker, template="plotly_dark")
fig.show()
```
